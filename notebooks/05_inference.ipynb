```python
import joblib
import pandas as pd

model = joblib.load("sleep_quality_model.pkl")
```

```python
model.feature_names_in_
```

```python
novo_paciente = pd.DataFrame(columns=model.feature_names_in_)
novo_paciente.loc[0] = 0  # inicializa tudo com zero

# Preenchendo dados num√©ricos
novo_paciente.at[0, "Age"] = 20
novo_paciente.at[0, "Sleep Duration"] = 8
novo_paciente.at[0, "Activity"] = 60
novo_paciente.at[0, "Stress Level"] = 2
novo_paciente.at[0, "Heart Rate"] = 78
novo_paciente.at[0, "Steps"] = 1000
novo_paciente.at[0, "Pressure_Systolic"] = 120
novo_paciente.at[0, "Pressure_Diastolic"] = 75

# One-Hot
novo_paciente.at[0, "Gender_Male"] = 1
novo_paciente.at[0, "BMI_Obese"] = 0
novo_paciente.at[0, "BMI_Overweight"] = 0
novo_paciente.at[0, "Occupation_Doctor"] = 1
```

```python
pred = model.predict(novo_paciente)
print(f"Qualidade de sono estimada: {pred[0]:.2f}")
```
